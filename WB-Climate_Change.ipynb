```python
#El presente documento empleará una base de datos sobre cambio climático en el Perú disponible en los data sets del Banco Mundial.
#Para más información sobre la base de datos revisar el siguiente enlace: https://data.humdata.org/dataset/a9684254-7d3e-46e0-ba0f-563d92a46181?force_layout=desktop

#Primero veamos los pandas
import pandas as pd
```

```python
#Ahora se importará la base de datos sobre las emisiones totales de gases de efecto invernadero para el país
data1=pd.read_csv('TF-Data.csv')
```

```python
#Ahora vamos a limpiar la base de datos
#Ordenando de menor a mayor
data1.sort_values("Year", inplace = True)
```

```python
#Ahora vamos a unir dos bases de datos. 
#La primera es sobre las emisiones totales de gases de efecto invernadero para el país
#La segunda es sobre las emisiones de CO2
import pandas as pd
df1 = pd.read_csv('TF-Data.csv')
df2 = pd.read_csv('TF-Data4.csv')
df_final = df1.merge(df2, left_on = 'Year', right_on = 'Year')
```

```python
#Ordenando de menor a mayor
df_final.sort_values("Year", inplace = True)
```

```python
#Observando la base
df_final
```

```python
#Ahora realizaremos una limpieza de la base de datos. Primero, reconocemos qué columnas tenemos.
df_final.columns
```

```python
#Ahora, vamos a cambiar algunos nombres
nuevos_nombres = {'Country Name': 'Country Name_x', 'Country ISO3' : 'Country ISO3_x'}
```

```python
df_final.rename(columns = nuevos_nombres, inplace = True)
```

```python
#Eliminando columnas repetitivas
df_final.drop(['Country Name_y', 'Country ISO3_y'], axis=1)
```

```python
#Ahora, realizaremos los gráficos que nos interesan.
import altair as alt
alt.Chart(df_final).mark_rect()
```

```python
alt.Chart(df_final).mark_line().encode(
    x = "Year")
```

```python
#Definimos un tema para las fuentes
def tema_graf():
    font = "Arial"
    labelFont = "Arial Narrow"
    sourceFont = "Arial Narrow"
    
    return {
        "config" : {
             "title": {'font': font, 
                       'fontSize': 15,
                       'fontWeight': 300,
                       'offset':0,
                       'anchor':'left'},
             "axis": {
                  "labelFont": font,
                  "titleFont": font,
                  "domainColor": "gray"
             },
             "header": {
                  "labelFont": font,
                  "titleFont": font
             },
             "legend": {
                  "labelFont": labelFont,
                  "titleFont": labelFont,
                  "symbolType": 'line',
                  "padding": 1
                  
             }
             ,
             'background': '#ffffff'             
        }
    }

alt.themes.register('mi_tema', tema_graf)
alt.themes.enable('mi_tema')
```

```python
#Realizamos un gráfico de línea para ver la evolución de los gases de efecto invernadero en el Perú. 
alt.Chart(df_final).mark_line().encode(
    alt.X('Year:O', axis=alt.Axis(title='Año')),
    alt.Y('Value_x:Q', axis=alt.Axis(title='kiloton de CO2 (equivalente)')),
     ).properties(
    title={
      "text": ["Emisiones de gases de efecto invernadero en el Perú (1970-2018)"], 
      "subtitle": ["Fuente: World Bank", "Categoría: Cambio climático"],
      "color": "Black",
      "subtitleColor": "Gray"
    })
```

```python
#Realizamos un gráfico de línea para ver la evolución de la emisión de CO2 en el Perú. 
alt.Chart(df_final).mark_area(color = 'green', 
                           opacity = 0.5, 
                           line = {'color':'darkgreen'}).encode(
    alt.X('Year:O', axis=alt.Axis(title='Año')),
    alt.Y('Value_y:Q', axis=alt.Axis(title='CO2 (kt)')),
     ).properties(
    title={
      "text": ["Emisiones de CO2 en el Perú (1970-2018)"], 
      "subtitle": ["Fuente: World Bank", "Categoría: Cambio climático"],
      "color": "Black",
      "subtitleColor": "Gray"
    })
```

```python
#Intento de un gráfico múltiple
base = alt.Chart(df_final.reset_index()).encode(x='Year:O')

alt.layer(
    base.mark_line(color='green').encode(y='Value_x:Q'),
    base.mark_line(color='purple').encode(y='Value_y:Q'),     
    ).properties(
    title={
      "text": ["Emisiones de gases de efecto invernadero y CO2 en el Perú (1970-2018)"], 
      "subtitle": ["Fuente: World Bank", "Categoría: Cambio climático", "Verde:emisiones de gases de efecto invernadero","Morado: emisiones de CO2"],
      "color": "Black",
      "subtitleColor": "Gray"
    })
```
